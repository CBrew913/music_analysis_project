```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Importing Top 10 data file
file_path = Path('data/top10s.csv')
top_songs_df = pd.read_csv(file_path, encoding = "ISO-8859-1")
top_songs_df.head()
```

```python
# Refining the dataset
top_songs_df = top_songs_df[['title', 'artist', 'year', 'top genre', 'pop']]
top_songs_df.head()
```

```python
# Checking count of rows containing NaN
top_songs_df.isnull().sum()
```

```python
top_songs_df = top_songs_df.drop_duplicates(['artist', 'title'])

top_songs_df
```

```python
#Exporting lyric data
top_songs_df.to_csv('data/popularity.csv', index=False)
```
